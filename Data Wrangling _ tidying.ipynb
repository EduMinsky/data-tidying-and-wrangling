#### Occurrence data are prone to several bias erros, such as, temporal, environmental, spatial or taxonomy. First we are going to deal with the temporal bias. We will exclude old data (year < 1960), since the majoraty of environmental data that people use for ecological, macroecological and biogeographical works use recent env data (>1960)
#### We are going to also create a dataframe with our old data just to keep the data. It migth be useful in the future

```python
old_data=data.query('year < 1960')
data=data.query('year>=1960')
data.head()
```

```python
print(data.year.unique(),'\n',old_data.year.unique())
#Great, now our main dataframe has only data from 1960 onwards
```

#### Now, we will keep only records with geographic information (latitude and longitude)

```python
data=data[(data['decimalLatitude'].notnull())&(data['decimalLongitude'].notnull())]
```

####  Now, let's take a closer look on the issue column. It has information about if the data is reliable or not. We do not want to keep records that has values like COORDINATE_REPROJECTION_SUSPICIOUS GEODETIC_DATUM_INVALID COORDINATE_ROUNDED COORDINATE_OUT_OF_RANGE COUNTRY_COORDINATE_MISMATCH COORDINATE_UNCERTAINTY_METERS_INVALID

##### We can see that some of the values has more than one issue.
#### (COORDINATE_ROUNDED;REFERENCES_URI_INVALID;OCCURRENCE_STATUS_INFERRED_FROM_INDIVIDUAL_COUNT;INSTITUTION_MATCH_FUZZY;COLLECTION_MATCH_FUZZY are an example of several issues that a single observation migth has (they are separeted by ;). So, apart from dealing with single issues, we also have to deal if several issues in a single observation!


```python
#First, let's create a dataframe with only the null values for issue:
null_issue=data[data['issue'].isnull()]
null_issue.reset_index(drop=True, inplace=True)
```

```python
#Now, we will create a dataframe with not null values:
data2=data[data['issue'].notnull()]
data2.reset_index(drop=True, inplace=True)
```

```python
#Now, we create a lambda function to deal with this problem
my_lambda = lambda a,b,c,d,e,f:~data2.issue.str.match(pat=a)&~data2.issue.str.match(pat=b) &~data2.issue.str.match(pat=c) &~data2.issue.str.match(pat=d) &~data2.issue.str.match(pat=e) &~data2.issue.str.match(pat=f)
data2=data2[my_lambda('COORDINATE_REPROJECTION_SUSPICIOUS','GEODETIC_DATUM_INVALID','COORDINATE_ROUNDED','COORDINATE_OUT_OF_RANGE','COUNTRY_COORDINATE_MISMATCH','COORDINATE_UNCERTAINTY_METERS_INVALID')]
data3=pd.concat([data2,null_issue])
data3.reset_index(inplace=True,drop=True)
```

#### Great! Now, we need to validate our data. We are going to check if the distribution data falls into the distribution polygon of each species. These type of polygons are created by specialists and it serves as a guide to check for data validity

#### First, we are going to transform our data into geodataframe. Later, we are going to call our polygon objects

```python
import geopandas
from shapely.geometry import Point
#Creating a column with the geometry necessary to transform into a geodataframe
data3['geometry'] = data3.apply(lambda x: Point((float(x.decimalLongitude), float(x.decimalLatitude))), axis=1)
#now, transforming into geodaframe
my_geo_data= geopandas.GeoDataFrame(data3)
type(my_geo_data)
```

```python
my_geo_data
```

```python
#Applying coordinate reference to our data. 4326 means WGS 1994
my_geo_data.set_crs(epsg=4326,inplace=True)
my_geo_data.crs
```

```python
#Reading polygon for each species:
my_geofiles=[]
for file in glob.glob('C:\\Users\\User\\Documents\\GitHub\\GBIF_Data Wrangling\\E_*.shp'):
     my_geofiles.append(geopandas.read_file(file))
```

```python
#Checking 1 species
import matplotlib.pyplot as plt
figure, axis = plt.subplots(nrows=1, ncols=1, figsize=(18,8))
my_geofiles[0].plot(ax=axis)
my_geo_data[my_geo_data['species']=='Epinephelus itajara'].plot(ax=axis,color='black')
#Let's retain only points that fall into the polygon
```

```python
#Creating separated objects for each species:
itajaraDF,itajara_poly=my_geo_data[my_geo_data['species']=='Epinephelus itajara'],my_geofiles[0]
marginatusDF,marginatus_poly = my_geo_data[my_geo_data['species']=='Epinephelus marginatus'],my_geofiles[1]
morioDF,morio_poly=my_geo_data[my_geo_data['species']=='Epinephelus morio'],my_geofiles[2]
```

```python
#Defining our function to deal with this problem. We are going to do a spatial join
def points_in_poly(geodata,poly):
    from geopandas.tools import sjoin
    a=sjoin(geodata, poly, how='left') #Applying the spatial join 
    
    my_points_poly=a[a['index_right'].notnull()]#retaining only observations that has not null values of the right index
        
    return my_points_poly
```

```python

itajara_inside_poly=points_in_poly(geodata=itajaraDF,poly=itajara_poly)
marginatus_inside_poly=points_in_poly(geodata=marginatusDF,poly=marginatus_poly)
morio_inside_poly=points_in_poly(geodata=morioDF,poly=morio_poly)
```

```python
print('\n','For Epinephelus itajara we have {} observations that were validated'.format(itajara_inside_poly.shape[0]),'\n','For Epinephelus marginatus we have {} observations that were validated'.format(marginatus_inside_poly.shape[0]),
     '\n','For Epinephelus morio we have {} observations that were validated'.format(morio_inside_poly.shape[0]))
```
